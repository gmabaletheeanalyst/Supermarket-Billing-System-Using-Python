```python
# Executes for every new customer
print("-" * 50)
print("Welcome to the Pythonfresh SuperMarket Billing System")
print("-" * 50)

# Name and Phone of the customer
name = input("Enter your name: ")
phone = input("Enter your phone number: ")

# List of items and their prices (name, price)
supermarket_items = [
    ("Apples", 8.0),
    ("Bananas", 6.5),
    ("Bread", 12.0),
    ("Eggs (Dozen)", 25.0),
    ("Milk (1L)", 10.0),
    ("Cheese", 35.0),
    ("Tomatoes", 9.0),
    ("Potatoes", 7.0),
    ("Onions", 6.0),
    ("Chicken (1kg)", 50.0),
    ("Beef (1kg)", 65.0),
    ("Rice (1kg)", 30.0),
    ("Pasta", 15.0),
    ("Cooking Oil (1L)", 40.0),
    ("Sugar (1kg)", 18.0),
    ("Salt (1kg)", 5.0),
    ("Butter", 20.0),
    ("Juice (1L)", 22.0),
    ("Cereal Box", 28.0),
    ("Tea Pack", 16.0)
]

# Convert the list to a dictionary so we can look up prices by name
products_items = dict(supermarket_items)

# Show available products
print("-" * 50)
print("Available Products and Prices:")
print("-" * 50)
for item, price in products_items.items():
    print(f"{item:<25} : ${price:.2f}")
print("-" * 50)

# Start billing
while True:
    amount = 0
    cart = {}
    discount = 0
    total_amount = 0

    while True:
        p_item = input("Enter product name you want to buy: ").title()
        
        if p_item not in products_items:
            print("Item not found. Please try again.")
            continue

        try:
            quantity = float(input(f"Enter quantity of {p_item}: "))
        except ValueError:
            print("Invalid quantity. Please enter a number.")
            continue

        price = products_items[p_item]
        total = price * quantity
        cart[p_item] = [price, quantity, total]
        amount += total

        repeat = input("Do you want to add more items? (yes/no): ").lower()
        if repeat != "yes":
            break

    # Calculate discount
    if amount <= 500:
        discount = 0.05
        Percent = "small basket discount 5%"
    elif amount <= 1500:
        discount = 0.10
        Percent = "middle saver discount 10%"
    elif amount <= 3500:
        discount = 0.15
        Percent = "weekly saver discount 15%"
    elif amount <= 5000:
        discount = 0.20
        Percent = "heavy basket discount 20%"
    else:
        discount = 0
        Percent = "no discount"

    total_amount = amount - (amount * discount)

    # Print receipt
    print("-" * 50)
    print("Name: ", name)
    print("Phone Number: ", phone)
    print("-" * 50)
    print("{:<20}{:<10}{:<10}{:<10}".format("Item", "Price", "Qty", "Total"))
    print("-" * 50)

    for item, values in cart.items():
        print("{:<20}${:<9.2f}{:<10.1f}${:<.2f}".format(item, values[0], values[1], values[2]))

    print("-" * 50)
    print(f"Final Amount: ${amount:.2f}")
    print(f"Discount Applied: {Percent}")
    print(f"Payable Amount: ${total_amount:.2f}")
    print("-" * 50)
    print("Thank you for shopping at PythonFresh Supermarket!")
    print("-" * 50)
    print("\n\n")
    break
```
